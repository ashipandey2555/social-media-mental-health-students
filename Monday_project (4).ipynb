```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load your dataset (change the file path to where your CSV is saved)
df = pd.read_csv("data.csv")

# Clean 'Origin_0': remove NaN, strip spaces, make consistent capitalization
df['Origin_0'] = df['Origin_0'].dropna().astype(str).str.strip().str.title()

# Remove rows where Origin_0 is missing after cleaning
df_clean = df.dropna(subset=["Origin_0"])

# Get top 10 origins by count
top_countries = df_clean['Origin_0'].value_counts().head(10).reset_index()
top_countries.columns = ['Origin_0', 'Count']

# Calculate average SCAS score for each origin
avg_scas = df_clean.groupby('Origin_0')['SCAS_TOTAL_0'].mean().reset_index()
avg_scas.columns = ['Origin_0', 'Avg_SCAS']

# Merge counts with averages
top_countries = top_countries.merge(avg_scas, on='Origin_0')

# Plot bar chart (color based on Avg_SCAS)
plt.figure(figsize=(12, 6))
sns.barplot(
    x='Count',
    y='Origin_0',
    data=top_countries,
    hue='Avg_SCAS',
    palette='coolwarm',
    dodge=False
)

# Add labels for Avg_SCAS
for i, row in top_countries.iterrows():
    plt.text(row['Count'] + 0.5, i, f"{row['Avg_SCAS']:.1f}", va='center')

plt.title("Top 10 Origins in My Dataset (Color = Avg SCAS Score)", fontsize=14, fontweight='bold')
plt.xlabel("Number of Students")
plt.ylabel("Origin (Country)")
plt.legend(title="Avg SCAS", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
